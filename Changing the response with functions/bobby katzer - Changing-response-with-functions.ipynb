## Function Fundamentals

Functions are little tools that we first define and then execute.  This is how we defined the `playerPage` function.

````
def playerPage():
  return "<h1> welcome to the player page </h1>"
````

### Defining a function

We define a function with the `def` keyword followed by our `name` of the function.  We can name our function anything so long as there are no spaces and it begins with a lowercase letter.  

After the function name, we need two parentheses (which we'll explain soon) followed by a colon to begin the body of the function.

```python
  def functionName():
    # function body
```

On the line after the function name, with indented code comes the body of the function.  Everything in the body of a function is locked within the walls of the function, except for the return statement which is catapulted over the wall.  We surround our HTML in quotes as the HTML is a **string**, which just means text. 

So those are the components of defining a function.
1. **The function name** - which consists of the `def` keyword, followed by `name():`

2. **The function body** which is what the function does, and ends with a `return` statement, to throw some data over the walls of the function

### Executing a function

After defining a function,  you can execute the code defined inside of the function by typing the name of the function followed by the parentheses.  

> In the gray box below, move your cursor to the top left and again press play.  (If you get an error, read the note directly below.)

````
playerPage()
````
